```python
from pathlib import Path
import pandas as pd
```

```python
PROCESSED_PATH = Path("../data/processed/hourly_demand_sample.parquet")

df = pd.read_parquet(PROCESSED_PATH)
print(df.shape)
df.head()
```

## create lag / sliding window

```python
df_rolling = df.copy()

for i in range(1, 25):
    df_rolling[f'lag_{i}'] = df_rolling['demand'].shift(i)

df_rolling = df_rolling.dropna().reset_index(drop=True)
```

```python
df_rolling.head()
```

```python
df_rolling.columns
```

```python
PROCESSED_PATH = Path("../data/processed/rolling_sample.parquet")

PROCESSED_PATH.parent.mkdir(parents=True, exist_ok=True)
df_rolling.to_parquet(PROCESSED_PATH, index=False)
```
