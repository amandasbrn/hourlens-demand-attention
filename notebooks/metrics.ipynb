```python
import pandas as pd
from pathlib import Path
```

```python
RF_METRIC = Path("../outputs/evaluation/random_forest_metrics.csv")
LSTM_BASELINE_METRIC = Path("../outputs/evaluation/lstm_baseline_metrics.csv")
LSTM_ATTN_METRIC = Path("../outputs/evaluation/lstm_attention_metrics.csv")
LSTM_ATTN_REG_METRIC = Path("../outputs/evaluation/lstm_attention_reg_metrics.csv")

RF_METRIC_DF = pd.read_csv(RF_METRIC)
LSTM_BASELINE_METRIC_DF = pd.read_csv(LSTM_BASELINE_METRIC)
LSTM_ATTN_METRIC_DF = pd.read_csv(LSTM_ATTN_METRIC)
LSTM_ATTN_REG_METRIC_DF = pd.read_csv(LSTM_ATTN_REG_METRIC)
```

```python
result = pd.concat([RF_METRIC_DF, LSTM_BASELINE_METRIC_DF, LSTM_ATTN_METRIC_DF, LSTM_ATTN_REG_METRIC_DF])
result
```

```python
SUMMARY_PATH = Path("../outputs/evaluation/metrics_summary.csv")

result.to_csv(SUMMARY_PATH, index=False)
```
